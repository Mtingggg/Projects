```python
import pandas as pd
import numpy as np
import datetime
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor

df = pd.read_excel(r'D:\Users\user\Desktop\碩班課程\資料探勘\HW2\107年新竹站_20190315.xls')
```

```python
##Q1
df=df.replace('NR',0)
df.shape
```

```python
for col in df.iloc[:,3:]:  #轉成numeric,非數字(文字或符號)用NA補
    df[col]=pd.to_numeric(df[col],errors='coerce')
df.head()
```

```python
df['日期'] = df['日期'].apply(lambda x: datetime.datetime.strptime(x,'%Y/%m/%d'))  #先轉成日期型態後面才能拿
df['月份'] = df['日期'].apply(lambda x: datetime.datetime.strftime(x,'%m')).astype('int64')   
df['日期'] = df['日期'].apply(lambda x: datetime.datetime.strftime(x,'%Y/%m/%d'))  #再把他轉回string
df.head()
```

```python
#發現 NO, NOx, NO2在7月有缺值
print(df.loc[:,'測項'].value_counts())
print(df.loc[df['測項']=='NO',:]['月份'].value_counts())
```

```python
# 補上他們的欄位
df.loc[6564]=list(['2018/07/15','新竹','NO'])+['NA']*25
df.loc[6565]=list(['2018/07/15','新竹','NO2'])+['NA']*25
df.loc[6566]=list(['2018/07/15','新竹','NOx'])+['NA']*25
df.loc[6567]=list(['2018/07/16','新竹','NO'])+['NA']*25
df.loc[6568]=list(['2018/07/16','新竹','NO2'])+['NA']*25
df.loc[6569]=list(['2018/07/16','新竹','NOx'])+['NA']*25
```

```python
df=df.sort_values(by=['日期','測項']).reset_index(drop=True)
df=df.drop(['月份'],axis=1)
```

```python
train_temp=[]
test_temp=[]
for i in range(18):     
    train_temp.append([])     
    test_temp.append([])
```

```python
for idx,col in enumerate(df['測項'].unique()):
    train_temp[idx]=df[df['測項']==col].iloc[:,3:].values.reshape(1,-1)
feature=pd.DataFrame(np.asmatrix(np.array(train_temp))).transpose()
feature.columns=[df['測項'].unique()]
feature.head()
```

```python
datelist=[]
timelist=[]
for date in df['日期'].unique():
        datelist.append([date]*24)
        timelist.append(list(range(0,24)))
feature['日期']=np.array(datelist).reshape(-1,1)  
feature['時間']=np.array(timelist).reshape(-1,1)  
feature.head()
```

```python
for col in feature.iloc[:,3:-2]:  #轉成numeric,非數字(文字或符號)用NA補     
    feature[col]=pd.to_numeric(feature[col],errors='coerce')
```

```python
dfleft=feature.ffill() #找前面的數字來補
dfright=feature.bfill() #找後面的數字來補
```

```python
for col in feature.iloc[:,:-2]:
    feature[col]=pd.concat([dfleft[col],dfright[col]],axis=1).mean(axis=1)
feature.head()
```

```python
#方法2，線性補值
for col in feature.iloc[:,3:-2]:  #轉成numeric,非數字(文字或符號)用NA補     
    feature[col]=pd.to_numeric(feature[col],errors='coerce')
df_fillna=feature.interpolate() #interpolate可以用前後數字線性補值
df_fillna.head()
```

```python
train=feature.loc[feature.iloc[:,18]>='2018/10/01'].reset_index(drop=True)
train.head()
```

```python
train_set=train.loc[train.iloc[:,18]<'2018/12/01']
test_set=train.loc[train.iloc[:,18]>='2018/12/01']
print(train_set.shape,test_set.shape)
```

```python
##第一題整理的形式
train.transpose().head()
```

```python
##Q2
def shift(data,num):
    return data[num:]
```

```python
train_X = pd.DataFrame()
temp=train_set['PM2.5']
for i in range(len(train)):
    if len(temp)>6:
        train_X=train_X.append(pd.DataFrame(temp.values[:6].reshape(1,6)),ignore_index=True)
        temp=shift(temp,1)
train_Y=train_set['PM2.5'][6:].reset_index(drop=True)
train_X.head()
```

```python
test_X = pd.DataFrame()
temp=test_set['PM2.5']
for i in range(len(train)):
    if len(temp)>6:
        test_X=test_X.append(pd.DataFrame(temp.values[:6].reshape(1,6)),ignore_index=True)
        temp=shift(temp,1)
test_Y=test_set['PM2.5'][6:].reset_index(drop=True)
test_X.head()
```

```python
##Q3
RF = RandomForestRegressor(n_estimators=500)
RF.fit(train_X.astype('int'), train_Y.astype('int').values.ravel())
RF_pred = RF.predict(test_X)
```

```python
OLS=LinearRegression()
OLS.fit(train_X.astype('int'), train_Y.astype('int'))
OLS_pred = OLS.predict(test_X)
```

```python
from sklearn.metrics import mean_absolute_error
print('RandomForest:',mean_absolute_error(test_Y, RF_pred))
print('OLS:',mean_absolute_error(test_Y, OLS_pred))
```
